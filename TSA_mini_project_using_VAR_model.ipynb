##STATIONARITY TESTS

````python
from statsmodels.tsa.stattools import adfuller
def adf_test(series, title=''):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] < 0.05:
        print(f'{title} is stationary')
    else:
        print(f'{title} is not stationary')


for col in df.columns[1:]:
  # Test for stationarity of EV sales
  adf_test(df[col], title=col)
````

````python
# First difference for most variables
df_diff1 = df[['DIESEL', 'petrol', 'hybrid', 'total']].diff(1).dropna()

# Second difference for EV and CNG
df_diff2_ev_cng = df[['ev', 'cng']].diff().diff().dropna()

# Combine the differenced data into a new DataFrame
df_stationary = pd.concat([df['Date'], df_diff1, df_diff2_ev_cng], axis=1).dropna()
````

````python
from statsmodels.tsa.stattools import adfuller
def adf_test(series, title=''):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] < 0.05:
        print(f'{title} is stationary')
    else:
        print(f'{title} is not stationary')


for col in df_stationary.columns:
  # Test for stationarity of EV sales
  adf_test(df_stationary[col], title=col)
````

````python
from statsmodels.tsa.stattools import grangercausalitytests
cols = ['DIESEL', 'petrol', 'cng', 'hybrid']
for col in cols:
  print(col)
  grangercausalitytests(df_stationary[['ev', col]], maxlag=4)

````

1. DIESEL
There is no Granger causality detected between DIESEL and ev up to 4 lags.
The p-values for all the lags are well above the significance threshold (typically 0.05), indicating that DIESEL does not Granger-cause ev.
2. Petrol
Granger causality is detected at lag 3.
At lag 3, the p-value for the F-test is 0.0176, which is below 0.05, indicating significant Granger causality.
The other test statistics (ssr-based chi2 and likelihood ratio test) also confirm this with low p-values.
At lag 4, the p-values approach significance, but they are slightly above 0.05 (e.g., 0.0738 for the F-test, 0.0502 for chi2 test), so we cannot confirm strong causality at lag 4.
3. CNG
There is strong Granger causality detected between CNG and ev across all lags.
At lag 1, the p-value for the F-test is 0.0052, which is significant.
At lag 2, the p-value drops to 0.0000, which indicates extremely strong causality.
The trend continues at lags 3 and 4 with p-values remaining at 0.0000, indicating that CNG is a significant Granger cause of ev.
4. Hybrid
Granger causality is detected at lag 3.
The p-value for the F-test is 0.0000, which is highly significant.
Granger causality is also detected at lag 4, with p-values for all tests being 0.0000, indicating strong causality at higher lags as well.
#Summary of Inferences:
###DIESEL does not appear to Granger-cause ev.
###Petrol Granger-causes ev at lag 3.
###CNG has strong causality for ev across all lags (1 to 4), with lag 2 showing the most robust causality.
###Hybrid vehicles Granger-cause ev starting at lag 3 and continuing strongly at lag 4.
*This suggests that CNG and Hybrid variables have a much stronger temporal influence on ev, while Petrol shows a moderate effect at lag 3. Diesel shows no significant influence.*

###VAR MODEL

````python
df_stationary.drop(columns = ['total'], inplace = True)
````

````python
# Convert the 'date' column to datetime format
df_stationary['Date'] = pd.to_datetime(df_stationary['Date'])

# Set the 'date' column as the index
df_stationary.set_index('Date', inplace=True)

# Verify the conversion
print(df_stationary.head())

df_stationary = df_stationary.asfreq('MS')
print(df_stationary.index.freq)
````

````python
df_stationary.drop(columns = ['DIESEL'], inplace = True)
````

````python
from sklearn.decomposition import PCA
pca = PCA(n_components=min(df_stationary.shape[1], 4))
data_pca = pd.DataFrame(pca.fit_transform(df_stationary), index=df_stationary.index)
````

````python
from statsmodels.tsa.api import VAR

# Fit the VAR model on differenced data
model = VAR(data_pca)

# Choose the optimal lag length using Akaike Information Criterion (AIC)
lag_order = model.select_order(maxlags=3)
print(lag_order.summary())

# Fit the model using the chosen lag length
var_model = model.fit(lag_order.aic)
print(var_model.summary())
````

````python
from statsmodels.stats.stattools import durbin_watson

# Compute Durbin-Watson statistics
dw_stats = durbin_watson(var_model.resid)

for col, stat in zip(df_stationary.columns, dw_stats):
    print(f'{col}: {stat}')
````

##3. VAR Model Estimation:
The VAR model was estimated with 3 lags.

````python
import pandas as pd
import matplotlib.pyplot as plt

def forecast_all_for_second_var(var_results, steps=12):
    try:
        # Get the original data (only the second variable)
        original_data = var_results.model.data.orig_endog
        second_var = original_data.iloc[:, 1]  # Assuming second variable is in the second column

        # Ensure the index is a DatetimeIndex
        if not isinstance(original_data.index, pd.DatetimeIndex):
            raise ValueError("The index must be a DatetimeIndex.")

        # Get in-sample predictions (fitted values)
        fitted_values = var_results.fittedvalues.iloc[:, 1]  # Fitted values for second variable

        # Generate out-of-sample forecast
        forecast_input = original_data.values[-var_results.k_ar:]  # Use the last 'k_ar' observations as input
        forecast = var_results.forecast(forecast_input, steps=steps)
        forecast_second_var = forecast[:, 1]  # Forecast for the second variable

        # Create a DataFrame for the out-of-sample forecast
        last_date = original_data.index[-1]
        forecast_index = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=steps, freq='MS')
        forecast_df = pd.DataFrame(forecast_second_var, index=forecast_index, columns=[second_var.name])

        # Combine in-sample and out-of-sample data
        all_values_df = pd.concat([fitted_values, forecast_df], axis=0)

        # Plot the original, fitted, and forecasted values for the second variable
        plt.figure(figsize=(12, 6))
        plt.plot(second_var.index, second_var, label='Observed', color='blue')
        plt.plot(fitted_values.index, fitted_values, label='Fitted (In-sample)', color='green')
        plt.plot(forecast_df.index, forecast_df, label='Forecast (Out-of-sample)', color='red')
        plt.title(f'Full Forecast for {second_var.name}')
        plt.legend()
        plt.tight_layout()
        plt.show()

        return all_values_df

    except Exception as e:
        print(f"An error occurred: {e}")

# Usage:
all_forecast_df = forecast_all_for_second_var(var_model)
print(all_forecast_df)
````

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_forecast_second_var(original_data, forecasted_data):
    try:
        # Extract only the second variable from the original data
        second_var = original_data.iloc[:, 1]  # Assuming second variable is in the second column

        # Ensure the original and forecasted data have the same indices
        common_index = second_var.index.intersection(forecasted_data.index)
        y_true = second_var.loc[common_index].values
        y_pred = forecasted_data.loc[common_index].values

        # Evaluation metrics for the second variable
        mae = mean_absolute_error(y_true, y_pred)
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_true, y_pred)

        # Print the results for the second variable
        print(f'MAE (2nd var): {mae}')
        print(f'RMSE (2nd var): {rmse}')
        print(f'R-squared (2nd var): {r2}')

    except Exception as e:
        print(f"An error occurred during evaluation: {e}")

# Usage:
original_data = var_model.model.data.orig_endog  # Original data from the VAR model
evaluate_forecast_second_var(original_data, all_forecast_df)  # Pass the full forecast DataFrame (including fitted and forecasted values)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def inverse_double_difference(original_series, diff_forecast):
    """
    Convert double-differenced forecasts back to original scale

    Parameters:
    original_series: Original (non-differenced) time series
    diff_forecast: Double-differenced forecasted values
    """
    # Get the last two values from original series for reversing double difference
    last_value = original_series.iloc[-1]
    second_last_value = original_series.iloc[-2]

    # Get the last difference (for first integration)
    last_diff = last_value - second_last_value

    # First integration (from second difference to first difference)
    first_diffs = [last_diff]  # Initialize with last known first difference
    for diff_value in diff_forecast:
        new_diff = first_diffs[-1] + diff_value
        first_diffs.append(new_diff)
    first_diffs = first_diffs[1:]  # Remove the initialization value

    # Second integration (from first difference to actual values)
    actual_values = [last_value]  # Initialize with last known value
    for diff_value in first_diffs:
        new_value = actual_values[-1] + diff_value
        actual_values.append(new_value)
    actual_values = actual_values[1:]  # Remove the initialization value

    return pd.Series(actual_values, index=diff_forecast.index)

# Convert index to datetime if it's not already
df.index = pd.to_datetime(df.index)

# Filter data from 2014 onwards
start_date = pd.to_datetime('2014-01-01')
original_ev = df['ev'][df.index >= start_date]

# Convert forecasted values to actual scale
actual_forecast = inverse_double_difference(original_ev, forecast_df['ev'])

# Create the visualization
plt.figure(figsize=(15, 8))

# Plot historical data
plt.plot(original_ev.index, original_ev,
         label='Historical Data',
         color='blue',
         marker='o',
         markersize=6,
         linewidth=2)

# Plot forecasted values
plt.plot(actual_forecast.index, actual_forecast,
         label='Forecasted',
         color='red',
         marker='*',
         markersize=8,
         linewidth=2,
         linestyle='--')

# Customize the plot
plt.title('EV Sales: Historical and Forecasted Values (2014 onwards)',
         fontsize=14,
         pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of EV Sales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add annotations for last historical point and first forecast
last_historical = original_ev.iloc[-1]
first_forecast = actual_forecast.iloc[0]

plt.annotate(f'Last Historical: {int(last_historical):,}',
            xy=(original_ev.index[-1], last_historical),
            xytext=(10, 10),
            textcoords='offset points',
            fontsize=10)

plt.annotate(f'First Forecast: {int(first_forecast):,}',
            xy=(actual_forecast.index[0], first_forecast),
            xytext=(10, -15),
            textcoords='offset points',
            fontsize=10)

# Shade the forecast region
plt.axvspan(actual_forecast.index[0], actual_forecast.index[-1],
            alpha=0.1,
            color='red',
            label='Forecast Period')

# Add a vertical line separating historical and forecast data
plt.axvline(x=original_ev.index[-1],
            color='gray',
            linestyle='--',
            alpha=0.5)

plt.tight_layout()
plt.show()

# Print summary statistics
print("\nSummary Statistics:")
print(f"Historical Data Range: {original_ev.index[0].strftime('%Y-%m')} to {original_ev.index[-1].strftime('%Y-%m')}")
print(f"Forecast Period: {actual_forecast.index[0].strftime('%Y-%m')} to {actual_forecast.index[-1].strftime('%Y-%m')}")
print(f"\nLast Historical Value: {int(last_historical):,}")
print(f"First Forecasted Value: {int(first_forecast):,}")
print(f"Forecasted Values Range: {int(actual_forecast.min()):,} to {int(actual_forecast.max()):,}")

# Print monthly forecasted values
print("\nMonthly Forecasted Values:")
for date, value in actual_forecast.items():
    print(f"{date.strftime('%Y-%m')}: {int(value):,}")
````
